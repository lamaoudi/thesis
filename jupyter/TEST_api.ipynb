```python
import json
import csv
import requests
import time
```

```python
##coordinates for Kigali
latitude = "-2.2417996081589897,"

longitude = "29.800809309442215"

#Rwanda radius approx. 4x10^6 
#test radius of 2 km, 2000 m 
radius = "2500"
bus_type = "school"

api_key = ""
```

```python
url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json?location='+latitude+longitude+'&radius='+radius+'&type='+bus_type+'&key='+api_key
```

```python
print(url)
```

```python
headers = {'content_type':'application/json', 'User-Agent':'lamaoudi'}
```

```python
def csv_write():
    csvfile = open('google_maps.csv', 'w')
    writer = csv.writer(csvfile, delimiter=',')
    writer.writerow(["name", "type", "vicinity", "lat", "long"])
```

```python
response = requests.get(url=url, headers=headers).json()
#writer = csv.writer(csvfile, delimiter=',')
```

```python
with open('google_maps.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile, delimiter = ",")
    
    writer.writerow(["name", "type", "lat", "long"])
    for obj in response['results']:
        writer.writerow([obj['name'],obj['types'],obj['vicinity'], obj['geometry']['location']['lat'],obj['geometry']['location']['lng']])
        
        while ('next_page_token' in response):
            url2 = url + '&pagetoken=' + response['next_page_token']
            time.sleep(5)
            response = requests.get(url=url2, headers=headers).json()
            for obj in response['results']:
                writer.writerow([obj['name'],obj['types'],obj['vicinity'], obj['geometry']['location']['lat'],obj['geometry']['location']['lng']])
                
        else:
            print(response)
            
    if __name__ == '__main__':
        csv_write()
```
