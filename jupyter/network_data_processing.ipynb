```python
exist_hv_sub =  gpd.read_file(pathto_network + fn_exist_hv_sub)

index_transmission = np.where((exist_hv_sub.Secondary_ == 0.) | (exist_hv_sub.Secondary_ == 110.) | (exist_hv_sub.Secondary_ == 220.))
sub1 = exist_hv_sub.loc[index_transmission[0], :]
index_sub1 = np.array(sub1.index)
index_transmission = index_sub1[np.where(sub1.Primary_kV != 30.0)]
transmission_substations = sub1.loc[index_transmission, :]
transmission_substations.to_file(pathto_data+'/outputs/transmission_agg.shp')

index_sub1 = np.append(index_sub1, 25)
hv_substations = exist_hv_sub.drop(index_sub1, axis = 0)
hv_substations.to_file(pathto_data+'/outputs/hv_agg.shp')

mv_substations = sub1.drop(index_transmission, axis = 0)

##### Interperting Processed Data 

print("Number of Substations: " , len(exist_hv_sub))
print("Number of On-the-Ground Transmission Substations: ", len(transmission_substations))
print("Transmission Line Primary Voltages: ", list(np.unique(transmission_substations.Primary_kV)))
print("Transmission Line Secondary Voltages: ", list(np.unique(transmission_substations.Secondary_)))
print("Number of On-the-Ground HV/MV Substations: ", len(hv_substations))
print("HV/MV Substation Secondary Voltages: ", list(np.unique(hv_substations.Secondary_)))
print("HV/MV Substation Primary Voltages: ", list(np.unique(hv_substations.Primary_kV)))
print("MV Substations Voltages: ", list(np.unique(mv_substations.Primary_kV)))
```
